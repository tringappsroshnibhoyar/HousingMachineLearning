```python
HouseDf.columns.values.tolist()
HouseDf= HouseDf.dropna(axis= 0, how='any')
```

```python
x_Data = HouseDf[['longitude',
 'latitude',
 'housing_median_age',
 'total_rooms',
 'total_bedrooms',
 'population',
 'households',
 ]]

```

```python
y_Data = HouseDf[ 'median_house_value']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_Data, y_Data , test_size = 0.25, random_state = 101)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
predictor = LinearRegression()
```

```python
predictor.fit(x_train, y_train)
```

```python
cpff_df = pd.DataFrame(predictor.coef_,x_Data.columns, columns=['Coefficient'])
```

```python
cpff_df
```

```python
predictions = predictor.predict(x_test)
```
